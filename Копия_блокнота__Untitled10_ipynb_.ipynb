```python
import numpy as np
import matplotlib.pyplot as plt

# X o'qi uchun qiymatlarni yaratish (-2pi dan 2pi gacha)
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Sinus funksiyasining qiymatlarini hisoblash
y = np.sin(x)

# Sinus funksiyasining grafikasini chizish
plt.plot(x, y, label='sin(x)', color='b')

# Yozuvlar va boshqalar
plt.title('Sinus Funksiyasining Grafigi')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.axhline(0, color='black',linewidth=0.5)  # X o'qi
plt.axvline(0, color='black',linewidth=0.5)  # Y o'qi
plt.grid(True)
plt.legend()

# Grafika ko'rsatish
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tasodifiy nuqtalar yaratish
# 1000 ta tasodifiy nuqta (x va y)
x = np.random.rand(1000)  # 0 va 1 orasida tasodifiy sonlar
y = np.random.rand(1000)

# Scatter grafikasini chizish
plt.scatter(x, y, color='blue', alpha=0.5, s=10)

# Yozuvlar va boshqalar
plt.title('Tasodifiy Nuqtalar - Scatter Grafik')
plt.xlabel('X o\'qi')
plt.ylabel('Y o\'qi')

# Grafikni ko'rsatish
plt.show()
```
